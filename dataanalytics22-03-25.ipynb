```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

a = np.array([1,2,3,4,5])
print(a)
print(len(a))
print(type(a))
print(a[0])
print(a[-4])
print(a.ndim)
a[0] = 11
print(a)
```

```python
#2 dimensional array
b = np.array([[1,2,3,4],[11,22,33,66]])
print(b)
print(len(b))
print(b[0,1])
print(b[1,1])
print(b.ndim)
```

```python
#3 dimensional array
a = np.array([[[1,2,3],[11,22,33],[44,55,52]]])
print(a)
print("nuber of dimensional",a.ndim)
print("Lenght of the array",len(a))
print("Lenght of inner array ",len(a[0]))
print(a[0,1,1])
a[0,1,1] = 44
print(a)
```
