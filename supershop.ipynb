Super store data analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('superstore.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# df.isnull().sum()
df.isna().sum()
```

```python
df.duplicated()
```

```python
s=df['Ship Mode'].unique()
s1=df['Segment'].unique()
s2=df['Country'].unique()
s3=df['State'].unique()
s4=df['Region'].unique()
s5=df['City'].unique()
print(s)
print(s1)
print(s2)
print(s3)
print(s4)
print(s5)
```

```python
#shifmode
s=df['Ship Mode'].value_counts()
print(s)
```

```python
plt.figure(figsize=(3,3))
c = ['green','blue','orange','red']
s.plot(kind='bar',color=c)
plt.ylabel('Product shiped Count')
plt.show()
```

```python
plt.figure(figsize=(3,3))
s=df.groupby('Ship Mode')['Sales'].sum()
s.plot(kind='bar')
plt.show()
```

```python
s.plot(kind='pie',autopct="%1.1f%%")
plt.show()
```

```python
s=df.groupby('Ship Mode')['Profit'].sum()
print(s)
```

```python
plt.figure(figsize=(2,2))
s.plot(kind='bar')
plt.ylabel('Profit')
plt.title('Profit base on ship mode')
plt.show()
```

```python
s.plot(kind='pie',autopct="%1.1f%%")
```
